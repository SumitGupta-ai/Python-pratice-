Level 1 Eassy Question

```python
class Car:
  def __init__(self, brand, model):
    self.brand = brand
    self.model= model

  def show(self):
    print(f"Brand {self.brand} Model {self.model}")


obj_1 = Car('Toyota','Fortuner')
obj_1.show()
```

```python
class Student:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def greet(self): # method
    print(f"Hello {self.name} and Your Age is {self.age}")

# object
s1 = Student('Sumit', 20)
s1.greet()
```

```python
class Mobile:
  def __init__(self, brand,price):
    self.brand = brand
    self.price = price

  def show(self):
    print(f"Model {self.brand} price {self.price}")

m1 = Mobile('Samsung', 12000)
m1.show()
```

Level 2 Medium Question

```python
class Rectangle:
  def __init__(self, length,width):
    self.length = length
    self.width = width

  def area(self):
    print(f" Rectangle Area is", self.length * self.width)

r1 = Rectangle(10,5)
r1.area()
```

```python
class Employee:
  company = 'ABC'
  def __init__(self, name, salary):
    self.name = name
    self.salary = salary

  def show_detail(self):
    print(f'Employee Name is : {self.name}\nand Salary : {self.salary}\nCompany Name : {self.company}')

emp = Employee('sumit', 12900)
emp.show_detail()
```

## Level 3 Inheritence

```python
class Animal:
  def eat(self):
    print('cat eats mouse')

class Dog(Animal):
  def eat(self):
    print("Dog eats cat")
s = Dog()
s.eat()
```

```python
class Animal:
  def food(self):
    print('Animal Eating Mans')

class Dog(Animal):
  pass

d = Dog()
d.food()
```
