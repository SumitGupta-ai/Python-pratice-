```python
# Q1. Write a Python program to swap two variables.
def swap(android, apple):
  return (f"{apple} {android}")

print(swap('vivo', 'iphone'))
```

```python
# Q2. Take user input and display it back to the user.
def user_input(name, age, course):
  return (f"Name: {name}\nAge: {age}\nCourse: {course}")


nm = (input('Enter a name: '))
ag = (input('Enter a age: '))
cou = (input('Enter a course: '))
print(user_input(nm, ag, cou))
```

```python
# Q3. Write a program to check if a number is even or odd.
def evenodd(n):
  if n % 2 == 0:
    return 'Even Number'
  else:
    return 'Odd Number'
print(evenodd(5))
```

```python
# Q4. Create a program that prints the multiplication table of a given number.
def multiplication(n):
  print('--Multiplication Table--')
  for i in range(1,11):
    product = f"{n} x {i} = {n*i}"
    print(product)

multiplication(7)
```

```python
# Q5. Write a program to find the largest of three numbers.
def largest_num(n=[2,3,4,10]):
  return max(n)
largest_num()
```

```python
# Q6. Convert temperature from Celsius to Fahrenheit.
def cels_to_fah(num=10):
  celsius = (num * 9/5 + 32)
  return celsius
cels_to_fah()
```

```python
# Q7. Write a program to calculate the factorial of a number using a loop.
def factorial(n=5):
  fact = 1
  for i in range(1, n+1):
    fact = fact * i
  return fact

factorial()
```

```python
# Q8. Create a program to count the number of vowels in a string.
def count_vowel(text):
  vowel = 'aeiou'
  count = 0
  for i in text:
    if i in vowel:
      count +=1
  return count
text = 'sumit is a cool guy'
print(f"Total vowel in text: {count_vowel(text)}")
```

```python
# Q9. Write a Python script to reverse a given string.
def revrese(text='sumit'):
  return text[::-1]
revrese()
```

```python
# wihout using slicing
def string_rev(text='python is good lang'):
  return "".join (reversed(text))
string_rev()
```

```python
# Q10. Check if a number is a palindrome.
def check_palindrome(text):
  if text == text[::-1]:
    print(f"{text}: Yah its Palindrome")
  else:
    print("NO its Not Palindrome")
text = 'mam'
check_palindrome(text)

```

```python
# Q11. Write a program to find the sum of first N natural numbers
def sum_of_n(n=6):
  sum = 0
  for i in range(1, n+1):
    sum = sum + i
  return sum

sum_of_n()
```
