```python
# Rivison oops
class Employee:
  def __init__(self, name, age, salary, company):
    self.name = name
    self.age = age
    self.salary = salary
    self.company = company

  def show_detail(self):
    print(f"Employee Name : {self.name}\nAge: {self.age}\nSalary: {self.salary}\ncompany: {self.company}")

emp = Employee('sumit', 20, 19000, 'TCS')
emp1 = Employee('mukesh', 22, 29000, 'TCS')
emp2 = Employee('rohan', 33, 17000, 'TCS')

emp.show_detail()
print()
emp1.show_detail()
print()
emp2.show_detail()
```

```python
#Check if a given year is a leap year or not.
def check_leap_year(year):
  if year % 400 == 0:
    print(f"{year} Leap Year")

  elif year % 100 == 0:
    print(f"{year} Not Leap Year")

  elif year % 4 == 0:
    print(f"{year} Leap Year")

  else:
    print(f"{year} Not leap year")

year = int(input('Enter a Year: '))

check_leap_year(year)
```

```python
# Create a program to print the Fibonacci series up to N terms.
f = int(input('Enter a Fibonacci Number: '))
a = 0
b = 1
for i in range(f):
  print(a, end=" ")
  c = a+b
  a = b
  b = c
```

```python
# Big Number provide fib that is slow exicute
def fib(n):
  if n <= 1:
    return n
  return fib(n-1) + fib(n-2)

num = 40
for i in range(num):
  print(fib(i), end=" ")
```

```python
n = 10
a,b = 0,1
count = 0
while count <= n:
  print(a, end=" ")
  c = a+b
  a = b
  b = c
  count += 1
```

```python
# Write a program to calculate the sum of digits of a number
def sum_of_digit(n):
  result = 0

  while n > 0:
    digit = n % 10
    result = result + digit
    n = n // 10
  return result

sum_of_digit(155)
```

```python
# Create a program to find the second largest number in a list
def second_largest(num=[25,400,75,100,200]):
  largest = num[0]
  second_lr = num[0]

  for n in num:
    if n > largest:
      second_lr = largest
      largest = n
    elif n > second_lr and n < largest:
      second_lr = n
  return second_lr
second_largest()
```
