```python
# Write a program to count the number of digits in an integer.
def count_digit(num):
  return len(str(num))
print(count_digit(1234))
print(count_digit(456549))
```

```python
# Loop method
def count_integer(num):
  count = 0

  while num != 0:
    num = num // 10
    count +=1

  return count
count_integer(4564)
```

```python
for i in range(1,1001):
  orgi = i
  temp = i
  sum = 0

  while temp != 0:
    digit = temp % 10
    sum = sum + digit**3
    temp = temp // 10
  if (sum == orgi):
      print(orgi)
```

```python
# Create a program to print all Armstrong numbers between 1 to 1000.
for i in range(1,1001):
    original = i
    temp = i
    sum = 0
    while temp != 0:
      digit = temp % 10
      sum = sum + digit**3
      temp = temp //10
    if (sum == original):
      print(f"Armstrong numbers: {original}")
```

```python
# Write a Python program to print a pattern of stars in a triangle.
n = 5
for i in range(n+1):
  for j in range(i):
    print("*", end="")
  print()

```

```python
# Create a calculator app using if-else
while True:
  user = input("Press Enter to continue or e to exit: ")
  if user == 'e':
    break
  number_1 = int(input('Enter a first number: '))
  number_2 = int(input('Enter a second number: '))
  operator = input("Choice Operator (+,-,*,%,/,//): ")
  if operator == "+":
    print(number_1 + number_2)
  elif operator == "-":
    print(number_1 - number_2)
  elif operator == "*":
    print(number_1 * number_2)
  elif operator == "%":
    print(number_1 % number_2)
  elif operator == "/":
    print(number_1 / number_2)
  elif operator == "//":
    print(number_1 // number_2)
  else:
    print("Invalid Operator")
```
