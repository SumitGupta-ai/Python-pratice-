```python
# 1. Write a function to check if a number is even.

def even_odd(num):

  if num == num//2*2:
    print(f"{num} even number")
  else:
    print(f"{num} Odd Number")

n = int(input('Enter a number: '))

even_odd(n)
```

```python
# 2. Create a list and find the sum of all its elements.
num = [25,50,75,100]
print(sum(num))
```

```python
num = [25,50,75,100]
total = 0
for n in num:
  total += n
print(total)
```

```python
num = [25,50,75,100]
total = 0
idx = 0
while idx < len(num):
  total += num[idx]
  idx += 1

print(total)
```

```python
# Write a program to find the maximum and minimum in a list.
# method 1
num = [25,50,75,100]
print(max(num))
print(min(num))
```

```python
# method 2
num = [125,50,75,100]
asd = num[::-1]
num.sort()
print(asd[0])
print(num[0])
```

```python
# method 3
num = [125,50,75,100,3040]
maximum = num[0]
minimum = num[0]
for i in num:
  if i > maximum:
    maximum = i
  elif i < minimum :
    minimum = i
print(maximum)
print(minimum)
```

```python
#  Create a program that removes duplicates from a list.
# method 1 using set
num = [30,34,23,43,32,30,43,32]
remove_duplicate = set(num)
print(list(remove_duplicate))
```

```python
# method 2
num = [30,34,23,43,32,30,43,32]
empty = []
for i in num:
  if i not in empty:
    empty.append(i)
print(empty)
```

```python
#  Write a function to reverse a list.
num = [25,50,72,100]
revrse = num[::-1]
print(revrse)
```

```python
# method 2
num = [25,50,72,100]
num.sort(reverse=True)
print(num)
```

```python
# method 3
num = [25,50,72,100]
empty = []
index = len(num) - 1

while index >= 0:
  empty.append(num[index])
  index -= 1
print(empty)
```
