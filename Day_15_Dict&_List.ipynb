```python
# Write a function to count the frequency of elements in a list.
num = [20,30,45,60,45,30]
freq = {}
for i in num:
  if i in freq:
    freq[i]+= 1
  else:
    freq[i] = 1
print(freq)
```

```python
#  Create a dictionary of squares of numbers from 1 to 10.
square_dict = {}
for i in range(2,11):
  square = i**2
  square_dict[i] = square
print(square_dict)
```

```python
cube_dict = {}
for i in range(2,11):
  cube_dict[i] = i**3
print("Number of Cube: ", cube_dict)
```

```python
# "Create a dictionary of even numbers..."
even_number = {}
for i in range(1,51):
  if i % 2== 0:
    even_number[i] = i
print(even_number)
```

```python
#  "Create a dictionary of factorials...
fact_dict = {}
for i in range(2,10):
  fact = 1
  for j in range(1, i+1):
    fact = fact * j
    fact_dict[i] = fact

print('Factorial Number:',fact_dict)



```
