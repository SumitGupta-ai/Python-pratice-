```python
# Write a function to find common elements in two lists.
def find_list(l1,l2):
  empty_list = []
  for i in l1:
    if i in l2:
      empty_list.append(i)
  return empty_list

l1 = [25,50,75,100]
l2 = [125,150,75,100]
print(find_list(l1,l2))
```

```python
# Write a function that returns the factorial of a number
def factorial(n):
  fact = 1
  for i in range(1, n+1):
    fact = fact * i
  return fact

n = 6
print(factorial(n))
```

```python
# Create a function that checks whether a string is a palindrome.
def palindrome(text):
  if text == text[::-1]:
    return('palindrome')
  else:
    return('no palindrome')

text = 'mam'
print(palindrome(text))


```

```python
# Write a function to count vowels in a string.
def count_vowel(text='sumit is good persOn'):
  vowel = 'aeiou'
  count = 0
  for t in text.lower():
    if t in vowel:
      count += 1
  return count
print(count_vowel())
```

```python
import string
def punch(text):
  result = ''
  for t in text:
    if t not in string.punctuation:
      result += t
  return result
text = 'sumit&@!#'
print(punch(text))
```
