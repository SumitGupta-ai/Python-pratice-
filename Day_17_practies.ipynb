```python
print('multiplication table')
try:
  n = int(input('Enter a number: '))
  for i in range(1,11):
    print(f"{n} x {i} = {n*i}")
except ValueError:
  print('Invalid Input')
```

```python
for i in range(1,1001):
  original = i
  temp = i
  sum = 0

  while temp != 0:
    n = temp % 10
    sum = sum + n**3
    temp = temp // 10
  if (sum == original):
    print(original)
```
