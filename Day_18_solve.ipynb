```python
# Q1. Write a Python program to check if a string has all unique characters.
def unique_char(text='sumit is good person'):
  unique_char = ''
  for char in text:
    if char in unique_char:
      return 'No! Not a Unique Char '
    else:
      unique_char += char
  return 'Yes! Yah its Unique Char'
unique_char()

```

```python
# Q2. Create a program that removes all duplicate characters from a string.
def remove_duplicate(text):
  new_str = ''
  for char in text:

    if char not in new_str:
      new_str += char
  return new_str

text = 'sumit is good boy'
print(remove_duplicate(text))
```

```python
# Q3. Write a script to count the frequency of each character in a string.
def freq(text):
  f = {}

  for char in text:
    if char in f:
      f[char] += 1

    else:
      f[char] = 1
  return f

text = 'sumit is good person'
print(freq(text.replace(" ",'')))
```

```python
from collections import Counter
text = 'sumit is good person'
f = Counter(text.replace(" ",''))
print(f)
```

```python
# Q4. Write a program that accepts a sentence and calculates the number of upper and lower case letters
def upper_lower_count(sen):
  is_upper = 0
  is_lower = 0
  for char in sen:
    if char.isupper():
      is_upper += 1
    elif char.islower():
      is_lower += 1
  return is_upper, is_lower

sen = 'Sumit Is'
upper_lower_count(sen)
```

```python
def count_upper_lower(sent):
  is_upper = 0
  is_lower = 0
  for char in sent:
    if 'A' <= char <= 'Z':
      is_upper += 1
    elif 'a' <= char <= 'z':
      is_lower +=1
  return is_upper, is_lower

sent = 'sumit is gOOD Person'
count_upper_lower(sent)

```
