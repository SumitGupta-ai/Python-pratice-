```python
#. Create a program to find the longest word in a sentence.
# method 1
def find_longest(sentence):
  word = sentence.split()

  large = max(word, key=len)
  return large
sentence = 'sumit is good person'
find_longest(sentence)
```

```python
#. Create a program to find the longest word in a sentence.
#mehtod 2
name = 'hello everyone that'
word = name.split()
longest_word = ''
for char in word:
  if len(char) > len(longest_word):
    longest_word = char
print(f"The longest word: {longest_word}")
```

```python
# find largest number
num = [2,34,5,5]
lo = num[0]
for i in num:
  if i > lo:
    lo = i
print(lo)
```

```python
# Q6. Write a program that takes a string and returns the string in reverse order without using [::-1].
def reverse_str(name):
  name = 'hello sumit'
  rev = ''
  for i in range(len(name) -1,-1,-1):
    rev += name[i]
  return rev

reverse_str(name)
```
