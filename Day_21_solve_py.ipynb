```python
# Q1. Write a Python program to check if a string has all unique characters.
def unique_char(sent):
  emp = ''
  for char in sent:
    if char in emp:
      return 'NO Its Not Unique Char'
    else:
      emp += char
  return 'YEs its unique char'

sent = input('Enter a sentence: ')
unique_char(sent)
```

```python
name = input('Enter a sentence: ')
words = name.split()
result = []
for word in words:
  new_word = word[0].upper() + word[1:]
  result.append(new_word)

final_sent = " ".join(result)
print(final_sent)


```

```python
# Q11. Create a program that extracts numbers from a string and returns their sum.
import re
n = 'sumit is good person 23 he is also 90'
total_num_str = re.findall(r'\d+', n)
total_sum = 0

for i in total_num_str:
  total_sum += int(i)

print(f'Total number of sum {total_sum}')
```

```python
def second_highest(n):
  largest_num = n[0]
  second_num = n[0]
  for i in n:
    if i > largest_num:
      second_num = largest_num
      largest_num = i
    elif i > second_num and i < largest_num:
      second_num = i
  return second_num
n = [29,38,40,28,19,30,8,90]
print(f"Second Heighest Number: {second_highest(n)}")
```
