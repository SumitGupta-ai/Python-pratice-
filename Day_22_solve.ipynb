```python
# Q20. Create a program to find the second largest number in a list.
# Eassy Method 1
n = [25,75,200,45,36,0]
n.sort()
slic = n[-2]
print(slic)
```

```python
n = [90,9,8,3,837]
lar = n[0]  # Start with 0 index lar value
lar = n[0]  # Start with 0 index sec value
sec = n[0]
for i in n:  # Loop
  if i > lar:
    sec = lar
    lar = i
  elif i > sec and i < lar:
    sec = i
print(sec)
```

```python
# 1. Write a function to check if a number is even.
def even_n(n):
    if n % 2 == 0:
      print(f'{n}Yes Even Number')
    else:
      print(f'{n} No Even Number')
n = int(input('Enter a number: '))
even_n(n)
```

```python
# 2. Create a list and find the sum of all its elements.
n = [10,20,30,40]
print(sum(n))
```

```python
# without sum function used
n = [10,20,30,40]
total = 0
for i in n:
  total += i
print(total)
```

```python
# third method
n = [10,20,30,40]
total = 0
i = 0
while i < len(n):
  total += n[i]
  i += 1
print(total)
```

```python
# 3. Write a program to find the maximum and minimum in a list.
num = [25,50,75,100,125,150,175]
print("maximum number", max(num))
print("Minimum number", min(num))
```

```python
num = [25,50,75,100,125,150,175]
max_num = num[0]
min_num = num[0]
for i in num:
  if i > max_num:
    max_num = i
  elif i < min_num:
    min_num = i
print("maximum number", max_num)
print("minimum number", min_num)
```
