```python
def multiplication(n):
  table = []
  for i in range(1,11):
    table.append(f"{n} x {i} = {n*i}")
  return '\n'.join(table)

n = int(input('Enter your multiplication number: '))
print(multiplication(n))
```

```python
# Q7. Write a program to calculate the factorial of a number using a loop.
def factorical(n):
  fact = 1
  for i in range(1, n+1):
    fact = fact*i
  return fact

n = 6
print(factorical(n))
```

```python
# Q8. Create a program to count the number of vowels in a string.
def count_vowel(text):
  count = 0
  vowel = 'aeiou'
  for i in text:
    if i in vowel:
      count += 1
  return count
text = input('enter a text: ')
count_vowel(text)
```

```python
# Q9. Write a Python script to reverse a given string.
name = input("Enter a name: ")
print(name[::-1])
```

```python
name = 'sumit'
reversed = ''
for i in range(len(name)-1,-1,-1):
  reversed += name[i]

print(reversed)
```
