```python
text = 'abracadabra'
str_to_list =  list(text)
str_to_list[5] = 'k'
text = ''.join(str_to_list)
print(text)
```

```python
def mutable_string(text, position, char):
  return text[:position] + char + text[position+1:]

text = 'abracadabra'

result = mutable_string(text,5, 'k')
print(result)
```

```python
def count_substring(text, sub_string):
  count = 0
  for i in range(len(text) - len(sub_string)+ 1):
    if text[i:i + len(sub_string) == len(sub_string)]:
      count +=1
  return count
  text = 'ABCDCDC'
  sub_string = 'CDC'

  count = count_substring(text, sub_string)
  print(count)
```
