```python
# Q10. Check if a number is a palindrome.
# Q11. Write a program to find the sum of first N natural numbers.
# Q12. Create a number guessing game.
# Q13. Write a program to print all prime numbers between 1 and 100.
```

```python
num = int(input("Enter a number: "))
original = num
reverse = 0

while num > 0:
  digit = num % 10
  reverse = reverse * 10 + digit
  num = num // 10


if original == reverse:
  print('Palindrom')
else:
  print('Not Palindrom')
```

```python
# Q11. Write a program to find the sum of first N natural numbers.
def sum_n(number):
  sum = 0
  for i in range(1, number+1):
    sum += i
  return sum

number = int(input('Enter a number: '))
print('The number of sum',number, 'Total number of sum',sum_n(number))
```

```python
# Q12. Create a number guessing game.
import random

print("Number Guessing Game")

while True:
  print("\nIf you'd like to play, please enter 'S'. To quit, enter 'Q'.")
  user = input('Enter S or Q: ').upper()

  if user == 'Q':
    print("Thanks for playing!")
    break
  elif user == 'S':
    computer = random.randint(1,50)
    while True:
      try:
        player = int(input("Enter your guess: "))
        if player > computer:
          print('Too High')
        elif player < computer:
          print('Too Low')
        else:
          print("You Win The Game! The number was", computer)
          break
      except ValueError:
        print("Invalid input. Please enter a whole number.")
  else:
    print("Invalid choice. Please enter 'S' to start or 'Q' to quit.")
```

```python
n = int(input("Enter a number: "))
if n<=1:
  print('not prime')
else:
  is_prime = True
for i in range(2, int(n**0.5)+1):
  if n % i == 0:
    is_prime = False
    break

if is_prime:
  print('prime')
else:
  print('Not Prime')
```

```python
# Alternetive method
n = int(input("Enter a number: "))
count = 0
for i in range(1, n+1):
  if n%i == 0:
    count += 1

if count == 2:
  print('prime')
else:
  print('not prime')
```
