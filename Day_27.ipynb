```python
num1 = int(input('Enter a First Number: '))
num2 = int(input("Enter a Second Number: "))
gcd = 1

if num1 < num2:
  small = num1
else:
  small = num2
for i in range(1,small + 1):
  if num1 % i == 0 and num2 % i == 0:
    gcd = i

print(gcd)
```

```python
import math
a = int(input('Enter a first number: '))
b = int(input('Enter a second number: '))

gcd = math.gcd(a,b)
lcm = (a*b) // gcd
print('gcd', gcd)
print('lcm', lcm)
```

```python
a = int(input('Enter a first number: '))
b = int(input('Enter a second number: '))
l = max(a,b)

while True:
  if l%a == 0 and l%b == 0:
    print('lcm', l)
    break
  l += 1
```

```python
# Q18. Check whether a character is a vowel or consonant.
text = 'a'
vowel = 'aeiou'
if text in vowel:
  print('This is Vowel Char')
else:
  print('This is consonant char')
```

```python
a = [90, 30, 40, 50, 29]

largest = float('-inf')
second_largest = float('-inf')

for i in a:
    if i > largest:
        second_largest = largest
        largest = i
    elif i > second_largest and i != largest:
        second_largest = i

print("Second Largest =", second_largest)
```
