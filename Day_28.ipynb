```python
num = array('i',[25,45,78])
for v in num:
  print(v,end=' ')
print('\n')
r = num[::-1]  # Reverse Array
for i in r:
  print(i, end=' ')
```

```python
def reverseArray(a):
    rev = a[::-1]
    for i in rev:
       print(i, end="")
n = int(input('Enter size: '))
a = []
for i in range(n):
  a.append(int(input()))
reverseArray(a)
```
