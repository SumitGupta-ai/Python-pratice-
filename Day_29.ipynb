```python
# count Even
arr = [10,5,20,8,4]
count = 0
for i in arr:
  if i % 2 == 0:
    count += 1
print(count)


```

```python
def ls(num, target):

  for index, elemnt in enumerate(num):
    if elemnt == target:
      return index
  return -1

num = [10,5,20,8,15]
print(ls(num, 8))
```

```python
# Second largest
arr = [10,5,20,8,15]
max_num = arr[0]
second_num = arr[0]

for i in arr:
  if i > max_num:
    second_num = max_num
    max_num = i
  elif i > second_num and i < max_num:
    second_num = i

print(second_num)
```
