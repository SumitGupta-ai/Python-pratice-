```python
def two_pointer(num, target):
  num.sort()
  left = 0
  right = len(num) - 1
  while left < right:
    total = num[left] + num[right]
    if total == target:
      return num[left], num[right]

    elif total > target:
      right = right - 1

    else:
      left = left + 1

  return  []

num = [1,20,4,5,77,8]

print(two_pointer(num, 12))
```

```python
# Two Pointer method solve reverse array
def reverse_arr(num):

  left = 0
  right = len(num) - 1

  while left < right:
    num[left],num[right] = num[right],num[left]
    left += 1
    right -=1
  return num
num = [25,50,75,100]
print(reverse_arr(num))


```

```python
# Two Pointer Palindrome
def palindrome(text):
  left = 0
  right = len(text) - 1

  while left < right:
    if text[left] != text[right]:
      return False
    else:
      left  += 1
      right -=1
  return True


text = 'racecar'
print(palindrome(text))
```
