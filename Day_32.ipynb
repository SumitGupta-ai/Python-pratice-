```python
def reverse_str(text):
  for i in range(len(text)-1,-1,-1):
    print(text[i], end='')

text = 'How are you'

reverse_str(text)
```

```python
# Two Pointer
def re(txt):
  text_list = list(txt)
  left = 0
  right = len(text_list) - 1
  while left < right:
    text_list[left], text_list[right] = text_list[right], text_list[left]
    left += 1
    right -= 1

  return "".join(text_list)

txt = 'how are you'
print(re(txt))
```

```python
# palindrome

def palindrome(num):
  original = num
  reverse = 0

  while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

  if original == reverse:
    return f"{original}, its Palindrome"
  else:
    return f"{original}, its not Palindrome"

# num = 121

print(palindrome(123))
print(palindrome(121))
print(palindrome(1001))
```
