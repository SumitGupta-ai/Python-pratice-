```python
# . Write a function that returns a new string made of every third character of the original string.
# simple method and without function
text = 'programmer'
for char in range(2, len(text), 3):
    print(text[char], end=" ")
```

```python
# method 2 with function
def third_char(text):
    new_string = ''
    for char in range(2, len(text), 3):
        new_string += text[char]
    return new_string
print(third_char('programmer'))
```

```python
# Write a program to find all palindromic substrings in a string
def find_polindromic(text):
    for i in range(len(text)):
        for j in range(i + 1, len(text) + 1):
            substring = text[i:j]
            if substring == substring[::-1]:
                print(substring)
find_polindromic('aba')
```
