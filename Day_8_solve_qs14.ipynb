```python
def check_even_odd(n):
  if n <= 0:
    return None
  if n % 2 == 0:
    return (f"{n} is even number")
  else:
    return (f"{n} is odd number")
n = int(input("Enter a number: "))
print(check_even_odd(n))
```

```python
def table(n):
  print(f"--Multiplication Table is-- {n}")
  for i in range(1,11):
    print(f"{n} x {i} = {n * i}")
n = int(input("Enter a number: "))
(table(n))
```

```python
# Formula: \(F = (C \times \frac{9}{5}) + 32\)
number = int(input('Enter a temprature: '))
formula = (number * 9/5) + 32
print(f"Celsius to Fahrenheit {formula} °F")
```

```python
def factorial(n):
  result = 1
  for i in range(1, n+1):
    result = result * i
  return result
n = int(input("Enter a number: "))
print(factorial(n))
```

```python
def count_vowel(text):
  count = 0
  vowel = 'aeiou'
  for char in text:
    if char.lower() in vowel:
      count +=1

  return count
text = input('Enter a sentence: ')
count_vowel(text)
```

```python
def reverse_string(text):
  text = text[::-1]
  return text
text = input('Enter a sentence: ')
reverse_string(text)
```

```python
def check_palindrome(text):
  p = text[::-1]
  if text == p:
    print('its palindrome')
  else:
    print('not palindrome')
text = input('Enter a sentence: ')
check_palindrome(text)
```

```python
def sum_n(number):
  sum = 0
  for i in range(1, number+1):
    sum = sum + i
  return sum
number = int(input('Enter a number: '))
sum_n(number)
```

```python
# Create a number guessing game.
import random
print('Number gussing game')
computer = random.randint(1,20)
while True:
  number = int(input('Enter your number: '))
  if number > computer:
    print('Its High Try again!')
  elif number < computer:
    print('Its Low Try again!')
  else:
    print('You win')
    break
```
