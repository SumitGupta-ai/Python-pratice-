```python
# Create a program to find the second largest number in a list.
number = [90,80,70,40,580,3,99]
largest_number = number[0]
second_largest = number[0]
for i in number:
  if i > largest_number:
    second_largest = largest_number
    largest_number = i
  elif i > second_largest and i < largest_number:
    second_largest = i
print(second_largest)
```

```python
# duplicate values
number = [99, 100, 120, 121, 120, 121]
large = float('-inf')
second = float('-inf')
for i in number:
  if i > large:
    second = large
    large = i
  elif i > second and i != large:
    second = i
print(f'largest number = {large}')
print(f"Second largest number = {second}")
```

```python
# Write a program to count the number of digits in an integer.
n = 12345
count = 0
while n > 0:
  n = n // 10
  count +=1
print(count)
```
