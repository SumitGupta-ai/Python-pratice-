```python
#. Write a function to flatten a nested list.
# without function

lst = [[1,2,3],[4,5,6]]
flatten = [item for i in lst for item in i]
print(flatten)
```

```python
# with function
def flatter(lst):
  flat = [item for i in lst for item in i]
  return flat
lst = [[1,2,3],[4,5,6]]
print(flatter(lst))
```

```python
num = [5,7,9,9,2,3,7]
unique = list(set(num))
if len(unique) < 2:
  print('no second hight')
unique.sort(reverse=True)
print(unique[1])
```

```python
def get_second_highest(numbers):
    unique_numbers = list(set(numbers))
    if len(unique_numbers) < 2:
        return None  # No second highest exists
    unique_numbers.sort(reverse=True)
    return unique_numbers[1]
num = [5,7,9,9,2,3,7]
print(get_second_highest(num))
```

```python
from builtins import sum as builtin_sum

def find_mising(num):
  n = len(num) + 1
  expeted = n * (n + 1) // 2
  acutal = builtin_sum(num)
  return expeted - acutal
number = [1, 2, 3, 4, 6,9]
print(find_mising(number))
```
