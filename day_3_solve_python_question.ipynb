```python
# 1. Write a function to check if a number is even.
# 2. Create a list and find the sum of all its elements.
#3. Write a program to find the maximum and minimum in a list.
#4. Create a program that removes duplicates from a list.
#5. Write a function to reverse a list.
```

```python
def evenodd(num):
    if num % 2 == 0:
      return 'its even'
    else:
      return 'its odd'
num = int(input('Enter a number: '))
print(evenodd(num))
```

```python
number = [2,3,4,5,6,7]
print(sum(number))
```

```python
number = [3,4,5,6]
print("its maximum number: ",max(number))
print("its minimum number: ",min(number))
```

```python
number = [7,8,9,9]
remove_dupli = (list(set(number)))
remove_dupli.sort()
print(remove_dupli)
```

```python
num = 'sumit'
revrse = num[::-1]
print(revrse)
```
