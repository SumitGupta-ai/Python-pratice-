```python
# Create a program to find the longest word in a sentence
sentence = input('Enter a sentence: ')
sentence_list = sentence.split()
longest_word= ''
for word in sentence_list:
  if len(word) > len(longest_word):
    longest_word = word
print(f"The longest words is {longest_word}")
```

```python
# . Write a program that takes a string and returns the string in reverse order without using [::-1]
text = 'sumit'
reversed = ''
for char in text:
  reversed = char + reversed
print(reversed)
```

```python
# Create a Python function to check if a string is a pangram.
def is_pangram(text):
  text = text.lower()
  unique_latter = set()
  for char in text:
    if char.isalpha():
      unique_latter.add(char)
  if len(unique_latter) == 26:
    return  True
  else:
    return False
text = "The quick brown fox jumps over the lazy dog"
print(is_pangram(text))
```

```python
# Write a Python script to sort words in a sentence alphabetically
sentence = "apple banana mangoes fruits"
word_split = sentence.split()
word_split.sort()
print(" ".join(word_split))
```

```python
#  Write a Python program to capitalize the first letter of each word in a sentence.
name = 'sumit gupta is a great boy'
capital_word = []
word = name.split()
for cap in word:
  capital_word.append(cap.capitalize())
print(' '.join(capital_word))
```
