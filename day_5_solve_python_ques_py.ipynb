```python
# Create a program that extracts numbers from a string without sum.

text = 'hello 420, and 90, samual'
print(''.join(char for char in text if (char.isdigit())))
```

```python
# Create a program that extracts numbers from a string and returns their sum.

import re
text = 'hello 420, and 90, samual'
number = [int(num) for num in re.findall(r'\d+', text)]
# print(number)
print(sum(number))
```

```python
# Write a program to replace all spaces in a string with underscores.

sentence = 'sumit is great boy'
undersocre = ''
for char in sentence:
  if char == " ":
    undersocre += "_"
  else:
    undersocre += char
print(undersocre)
```

```python
#Write a script to convert a string into title case without using .title().
text = 'sumit gupTa'
text_split = text.split()
for word in text_split:
  print(word.capitalize(), end=' ')
```
