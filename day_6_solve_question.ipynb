```python
# Create a program to filter out all non-alphabetic characters from a string
def noaplha(text):
  new_str = ''
  for i in text:
    if i.isalpha():
      new_str += i
    elif not  i.isalpha():
      new_str += ''
  return new_str
text = '20055 in the class'
print(noaplha(text))
```

```python
# Write a function that returns True if a string ends with a given suffix.

def checksuffix(start, suffix):
    if start.endswith(suffix):
      return True
    else:
      return False
start = 'happy word'
suffix  = 'word'
print(checksuffix(start, suffix))
```

```python
# Write a program that accepts a string and counts vowels and consonants.
text = 'python is interpreter language'
vowel_count = 0
const_count = 0
vowel = 'aeiou'
for vow in text.lower():
  if vow.isalpha():
    if vow in vowel:
      vowel_count += 1
    else:
      const_count += 1
print("vowerl Number: ", vowel_count)
print("constant Number: ", const_count)
```

```python
# Create a script to convert binary string to decimal
def binary_decimal(binary):
  result = 0
  for digit in binary:
    result = result * 2 + int(digit)
  return result
binary = '101101'
print(binary_decimal(binary))
```

```python
# Write a program to count the number of words starting with a vowel in a string.

text = 'python is interpreter language and'
count = 0
for word in text.split():
  if word[0].lower() in 'aeiou':
    count += 1

print('First charecter of vowel count :', count)
```

```python
# Create a script that takes a sentence and removes all stop words

sentence = 'sumit is great boy and he is also played cricket'
stop_word = {'is', 'am', 'are', 'the', 'in', 'he','and'} # manually creat stop words
result = ''
for word in sentence.split():
  if word not in stop_word:
    result += word + ' '
print(result)
```

```python
# Write a Python program to split a sentence into words and reverse each word

sentence = 'sumit is great boy and he is also played cricket'
for word in sentence.split():
  print(word[::-1],end=' ')
```
