```python
def evenodd(n):
    if n % 2 == 0:
        return(f'{n} is even number: ')
    else: 
        return('odd number: ')
    return n
print(evenodd(6))
    
```

```python
li = [2,3,4,5,6]
sum = 0
for i in li:
    sum += i
print(f"Total sum of number: {sum}")
```
