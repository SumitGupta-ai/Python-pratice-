```python
# 8. Create a dictionary and iterate over its keys and values
dic = {"a":1 , "b": 2}
for key, value in dic.items():
  print(key, value)
```

```python
import string
text = 'sumit is great boy @ and he is learning everything /<>;'
remove_punc = str.maketrans('','',string.punctuation)
remove_punc = text.translate(remove_punc)
print(remove_punc)
```

```python
#list com
text = 'sumit&'
clean = "".join(char for char in text if char not in string.punctuation)
print(clean)
```

```python
def capital(text):
  capi = text.title()
  return capi
text = 'sumit is good person'
print(capital(text))
```
