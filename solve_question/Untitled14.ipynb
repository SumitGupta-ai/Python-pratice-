```python
# Write a class “Calculator” capable of finding square, cube and square root of a number.
class Calculator:
  def __init__(self, n):
    self.n = n
  def sqaure(self):
    return (f"The square is {self.n*self.n}")
  def cube(self):
    return (f"The cube is {self.n*self.n*self.n}")
  def squareroot(self):
    return (f"The squareroot is {self.n**1/2}")
s = Calculator(5)
print(s.sqaure())
print(s.cube())
print(s.squareroot())
```
